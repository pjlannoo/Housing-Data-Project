- Getting rid of outliers based of grade
- Our target audience is for the typical average house buyer
- So no mansions/poor/luxury grades

```python
data = pd.read_csv('../../data/clean_kc_house_data.csv')
```

```python
data = data.drop(columns='Unnamed: 0')
```

```python
rel_grade = ['7 Average', '8 Good', '9 Better', '6 Low Average', '10 Very Good']

mask = data['grade'].isin(rel_grade)
```

```python
rel_data = data[mask]
```

- filtering by bedrooms

```python
rel_data = rel_data[rel_data['bedrooms'] < 7]
```
